```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("PremierLeagueSeason2024.csv")
```

```python
df.head(10)
```

```python
df.describe()
```

```python
df.sort_values(by='points', ascending=False, inplace=True)
plt.figure(figsize=(10, 6))
plt.bar(df['team'], df['points'], color='skyblue')
plt.title('Premier League Overview')
plt.xlabel('Points')
plt.ylabel('Teams')
plt.xticks(rotation=90)
plt.show()
```

```python
top_teams = df.iloc[:7]
top_teams
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(top_teams['goals_scored'], top_teams['goals_conceded'], color='blue', s=100)

for i, team in enumerate(top_teams['team']):
    plt.text(top_teams['goals_scored'][i] + 0.5, top_teams['goals_conceded'][i] + 0.5, team, fontsize=9)

plt.title('Goals Scored vs Goals Conceded by Team')
plt.xlabel('Goals Scored')
plt.ylabel('Goals Conceded')
plt.grid(True)

plt.show()
```

This scatter plot shows the relationship between **Goals Scored** and **Goals Conceded** for Top 6  teams. 

-**Manchester City** has scored the most goals and conceded the fewest, indicating they have an outstanding offense and defense. They are the most balanced team in terms of both scoring and preventing goals.

-Mid-tier teams like **Manchester United** and **Aston Villa** can score high amount of goals but struggle due to their defenses. This suggests that to compete at the top, a balanced approach between scoring and defending is essential.

```python
df['total_games'] = df['wins'] + df['draws'] + df['losses']
df['win_percentage'] = (df['wins'] / df['total_games'] ) * 100 
df['draw_percentage'] = (df['draws'] / df['total_games']) * 100 
df['losses_percentage'] = ( df['losses'] / df['total_games'] ) * 100

df.head(10)
```

```python
top_teams = df.iloc[:7]
top_teams
```

```python
top_teams = top_teams.set_index('team', drop=True)
```

```python
top_teams.head()
```

```python
top_teams[['wins', 'draws','losses']].plot(kind='bar', figsize=(10, 6), color=['skyblue', 'pink','orange'])

plt.title('Wins, Draws, Losses of Top 7 Teams')
plt.ylabel('Values')
plt.xlabel('Teams')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
```

```python
top_teams[['win_percentage', 'draw_percentage', 'losses_percentage']].plot(kind='line', figsize=(10, 6), marker='o', color=['skyblue', 'pink','orange'])

plt.title('Win, Draw, and Losses Percentages of Top 7 Teams')
plt.ylabel('Percentage (%)')
plt.xlabel('Teams')
plt.xticks(rotation=45)
plt.grid(True)
# Show the plot
plt.tight_layout()
plt.show()
```

-Both plots shows that the **Manchester City** is a dominant team compared to others. The gap between the **win percentage** and **losses percentage** also confirms they have consistently been the beest team, winning the majority of their games and rarely loosing.

-**Arsenal and Liverpool** are also strong opponents, but they have less consistency in converting games into wins. Their high draw rates indicate they drop points in matches they could potentially win.

-**Aston Villa** has the weakest performacnce metrics across both visuals, with low wins. They are also clearly lagging behind in terms of performance compared to other 6 teams.

```python

fig, axs = plt.subplots(3, 3, figsize=(15, 10))
axs = axs.flatten() 
for i, (team_name, row) in enumerate(top_teams.iterrows()):
    wins = row['wins']
    draws = row['draws']
    losses = row['losses']


    sizes = [wins, draws, losses]
    labels = ['Wins', 'Draws', 'Losses']
    colors = ['skyblue', 'pink','orange']  # Custom colors

    axs[i].pie(sizes, labels=labels, colors=colors, startangle=90)
    axs[i].axis('equal')  # Equal aspect ratio ensures pie is drawn as a circle.
    axs[i].set_title(f'{team_name} Performance')

    # Hiding  unused subplots 
for j in range(i + 1, len(axs)):
    axs[j].axis('off')

# Adjus layout
plt.tight_layout()
plt.show()
```
